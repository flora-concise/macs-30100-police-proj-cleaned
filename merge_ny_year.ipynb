```python
import pandas as pd

dfs = [pd.read_csv(f"ACSST5Y20{i}.S1501-Data.csv").iloc[1:] for i in range(13, 24)]


common_columns = list(dfs[0].columns)
for df in dfs[1:]:
    common_columns = [col for col in common_columns if col in df.columns]


for i in range(len(dfs)):
    dfs[i] = dfs[i][common_columns]


final_df = pd.concat(dfs, axis=0, ignore_index=True)
```

```python
dfs = [pd.read_csv(f"ACSST5Y20{i}.S1903-Data.csv").iloc[1:] for i in range(13, 24)]

common_columns = list(dfs[0].columns)
for df in dfs[1:]:
    common_columns = [col for col in common_columns if col in df.columns]


for i in range(len(dfs)):
    dfs[i] = dfs[i][common_columns]


final_df2 = pd.concat(dfs, axis=0, ignore_index=True)
```

```python
final_df2.shape
```

```python
final_df2.head()
```

```python
merged_df_ny = pd.merge(final_df, final_df2,
                    on='GEO_ID', 
                    how='inner')
```

```python
merged_df_ny.shape
```

```python
merged_df_ny.to_csv('ny_census.csv', index=False)
```
